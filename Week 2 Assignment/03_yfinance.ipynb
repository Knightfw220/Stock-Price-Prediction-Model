### Function Specification: `download_historical_data`

Implement the function `download_historical_data` to fetch historical price data using Yahoo Finance as the data source. This function should be capable of fetching historical data for a specified stock symbol between given start and end dates. Additionally, it should support an optional parameter for the data timeframe with a default value of `'1d'` (daily).

---

### Function Specifications

**Parameters:**
- `symbol`: The ticker symbol of the stock (e.g., `'RELIANCE.NS'`).
- `start_date`: Start date for the data in `'YYYY-MM-DD'` format.
- `end_date`: End date for the data in `'YYYY-MM-DD'` format.
- `timeframe`: The frequency of the data (`'1d'`, `'1wk'`, `'1mo'`), default is `'1d'`.

**Return:**  
A pandas `DataFrame` containing the fetched data.


```python
import yfinance as yf
def download_historical_data(symbol, start_date, end_date, timeframe='1d'):
  data=yf.download(symbol,start=start_date,end=end_date,interval=timeframe)
  return pd.DataFrame(data)
print(download_historical_data('AAPL','2012-1-1','2025-1-1','1mo')['Close'])
```

### Visualization

Create a function for plotting the stock closing prices. This function should accept parameters for the plot and its objective is to display the graph.
   


```python
def plot_stock_closing_prices(symbol, start_date, end_date, timeframe='1d'):
  data=download_historical_data(symbol, start_date, end_date, timeframe)
  df=pd.DataFrame(data)
  fig=plt.figure()
  ax=fig.add_axes([0,0,1,1])
  ax.plot(df['Close'])
  ax.set_xlabel('Date')
  ax.set_ylabel('Closing Price')
  ax.set_title(f'{symbol} Stock Closing Prices')

plot_stock_closing_prices('AAPL','2012-1-1','2025-1-1','1mo')
```

### Technical Analysis

**Simple Moving Averages (SMA) and Exponential Moving Averages (EMA)**  
   - Plot the SMA and EMA of 5 days of each of the three stocks that you have selected.
   - SMA and EMA graph should be plotted on the same graph
   - Hence you have to plot three graphs of each stock with SMA and EMA
    

```python
def plot_SMA_EMA(symbol, start_date, end_date, timeframe='1d'):
    data = download_historical_data(symbol, start_date, end_date, timeframe)
    df = pd.DataFrame(data)

    # Calculate SMA
    df['SMA_5'] = df['Close'].rolling(window=5).mean()

    # Calculate EMA
    df['EMA_5'] = df['Close'].ewm(span=5, adjust=False).mean()

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(df['Close'], label='Closing Price')
    plt.plot(df['SMA_5'], label='SMA 5')
    #plt.plot(df['EMA_5'], label='EMA 5')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'{symbol} - SMA and EMA (5 days)')
    plt.legend()
    plt.grid(True)
    plt.show()


# Plot for each stock
plot_SMA_EMA('AAPL', '2012-1-1', '2022-1-1', '1d')
plot_SMA_EMA('MSFT', '2012-1-1', '2022-1-1', '1d')
plot_SMA_EMA('AMZN', '2012-1-1', '2022-1-1', '1d')
```

**MACD**  
   - Plot MACD and Signal line for each of the three stocks.
   - Fast Length (Short-term EMA): 12 periods
   - Slow Length (Long-term EMA): 26 periods
   - Signal Line (Smoothing EMA): 9 periods

    

```python
def calculate_macd(data, fast_length=12, slow_length=26, signal_length=9):
    """Calculates MACD and Signal line."""
    short_ema = data['Close'].ewm(span=fast_length, adjust=False).mean()
    long_ema = data['Close'].ewm(span=slow_length, adjust=False).mean()
    macd_line = short_ema - long_ema
    signal_line = macd_line.ewm(span=signal_length, adjust=False).mean()
    return macd_line, signal_line

# Stock tickers
stock_tickers = ['AAPL', 'MSFT', 'AMZN']

for ticker in stock_tickers:
    data = download_historical_data(ticker, '2012-1-1', '2022-1-1', '1d')
    macd_line, signal_line = calculate_macd(data)

    plt.figure(figsize=(12, 6))
    plt.plot(macd_line, label='MACD')
    plt.plot(signal_line, label='Signal Line')
    plt.title(f'{ticker}')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()
```

**RSI**  
   - Plot RSI for each of the three stocks
   - Period = 14 days
   - Also show the overbought and oversold regions
   - Overbought condition: RSI above 70
   - Oversold condition: RSI below 30
   
    

```python
def calculate_rsi(data, period=14):
    delta = data['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Stock tickers
stock_tickers = ['AAPL', 'MSFT', 'AMZN']

for ticker in stock_tickers:
    data = download_historical_data(ticker, '2012-1-1', '2022-1-1', '1d')
    rsi = calculate_rsi(data)

    plt.figure(figsize=(12, 6))
    plt.plot(rsi)
    plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
    plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
    plt.title(f'{ticker} RSI (14 days)')
    plt.xlabel('Date')
    plt.ylabel('RSI')
    plt.legend()
    plt.grid(True)
    plt.show()
```
